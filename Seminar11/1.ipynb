```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
```

```python
x = Symbol('x')
```

Дана функция

```python
f1 = 5*x**2 + 10*x - 150
f1
```

1. Определить корни 

```python
x1, x2 = solve(f1)
print('Первый корень  = ', x1.evalf())
print('Второй корень  = ', x2.evalf())
```

Вычислили производную первого порядка исходной функции

```python
f2 = diff(f1)
f2
```

2 и 3 Найти интервалы, на которых функция возрастает и убывает

```python
fun1 = plot(f2,(x,-2,0), show=False)
fun1.show()
print(f'Функция убывает от -ꝏ до {extremum} и функция возрастает с {extremum} до +ꝏ')
```
